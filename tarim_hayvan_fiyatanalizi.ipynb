```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

buyuk = pd.read_excel (r'buyukbas.xls')
kucuk = pd.read_excel (r'kucukbas.xls')
```

Okunan büyükbaş verilerinin sütünlarına anlaşlır olması için isim verildi ve NaN içeren sütun ve satırlar kaldırıldı.

```
buyukbasFiyatlar = buyuk[3:14]
buyukbasFiyatlar.index=range(0,11)
columnnames = ["Buyukbas","2008","2009","2010","2011","2012","2013","2014","2015","2016","2017","2018","2019"]
buyukbasFiyatlar.columns = columnnames
buyukbasFiyatlar = buyukbasFiyatlar[pd.notnull(buyukbasFiyatlar["Buyukbas"])]
# buyukbasFiyatlar["Buyukbas"]
buyukbasFiyatlar
```

Büyükbaş hayvanlar için yıllara göre fiyatını görebileceğimiz bir tablo oluşturuldu. İlk olarak büyükbaşları almayacak şekilde datafraemi dolaşıyoruz ve sütun isimlerini bir listeye daha sonra tabloda isim vermek için atıyoruz. Verilerimizin içinde boşluklar olduğu için ve değerlerde geçersiz '-' olduğu için ilk önce sütunları string değerlere çeviriyoruz sonra boş değerleri 0 a çeviriyoruz. Değerlerimizi 0 yaptıktan sonra floata geri döndürüp çizdiriyoruz ve plotla isimlendiriyoruz.


```
plt.figure(figsize=(30,15))
plt.ylabel("Yıllara Göre Ortalama Fiyat")
plt.xlabel("Hayvanlar")
hayvanlar = buyukbasFiyatlar.Buyukbas
sutunlar = []
for column in (buyukbasFiyatlar.columns):
    if column is not "Buyukbas":
        sutunlar.append(column)
        a = buyukbasFiyatlar[column].astype(str)
        a =a.replace(to_replace='-',value = '0')
        plt.plot(hayvanlar,a.astype(float))
plt.legend(sutunlar)
# # ortalamaYillikFiyatlar = buyukbasFiyatlar.mean(axis = 0, skipna = True)
# # ortalamaYillikFiyatlar
```
